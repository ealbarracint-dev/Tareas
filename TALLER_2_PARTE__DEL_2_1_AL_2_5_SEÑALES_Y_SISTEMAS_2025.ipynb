Se deja como ejercicio la evaluación de las **integrales de convolución** para los tres casos analizados.  
Se debe determinar el resultado de $( y(t) = (x*h)(t) $) considerando el desplazamiento de una señal sobre la otra y analizar en qué intervalos existe superposición entre ellas.

> En el primer caso no hay superposición, es decir, $( y(t)=0 $) para $( t<0 $).


- Evaluar paso a paso la integral de convolución de dos señales rectangulares causales.  
- Identificar los tres casos:  
  1. Sin traslape (antes del contacto)  
  2. Con traslape (durante la intersección)  
  3. Sin traslape (después de que se separan)  
- Verificar que el resultado analítico corresponde a la forma triangular esperada.

**Señales consideradas:**  
$[
x(t) = u(t) - u(t - T_1), \qquad
h(t) = u(t) - u(t - T_2)
$]

donde $(u(t)$) es la función escalón unitario, y $(T_1, T_2 > 0$) son los anchos de cada rectángulo.

**Definición de la convolución:**
$[
y(t) = \int_{-\infty}^{\infty} x(\tau)\,h(t-\tau)\,d\tau
$]

**Resultado esperado (forma general):**
$[
y(t) =
\begin{cases}
0, & t < 0, \\
t, & 0 \le t \le T_{\min}, \\
T_{\min}, & T_{\min} \le t \le T_{\max}, \\
T_1 + T_2 - t, & T_{\max} \le t \le T_1 + T_2, \\
0, & t \ge T_1 + T_2,
\end{cases}
$]
donde $(T_{\min} = \min(T_1, T_2)$) y $(T_{\max} = \max(T_1, T_2)$).

- En el **primer caso** (sin traslape, $(t<0$)) no hay área común entre las señales → $(y(t)=0$).  
- En el **segundo caso**, la integral crece o se mantiene constante según la superposición.  
- En el **tercer caso**, al separarse las señales, la integral decrece hasta volver a cero.  
El resultado es una **forma triangular** que representa el área de intersección variable entre las señales.

### 1) Planteamiento
Trabajo con dos rectángulos **causales** unitarios (altura 1), de anchos $(T_1$) y $(T_2$):
$[
x(t)=u(t)-u(t-T_1),\qquad h(t)=u(t)-u(t-T_2).
$]
La convolución es:
$[
y(t)=(x*h)(t)=\int_{-\infty}^{\infty} x(\tau)\,h(t-\tau)\,d\tau.
$]
Como ambas señales valen 1 sólo en sus ventanas, la integral **se reduce a medir el largo del traslape** entre los intervalos:
$[
\tau\in[0,\,T_1]\quad \cap \quad \tau\in[t-T_2,\,t].
$]

---

### 2) Soporte de cada término
- $(x(\tau)=1$) si $(\tau\in[0,\,T_1]$), y 0 fuera.
- $(h(t-\tau)=1$) si $(t-\tau\in[0,\,T_2]\ \Leftrightarrow\ \tau\in[t-T_2,\,t]$); 0 fuera.

Entonces:
$[
y(t)=\int_{\max(0,\,t-T_2)}^{\min(T_1,\,t)} 1\; d\tau
= \big[\min(T_1,t)-\max(0,t-T_2)\big]_+,
$]
donde $([z]_+=\max(z,0)$) asegura que, si no hay traslape, el resultado sea 0.

---

### 3) Análisis por tramos (los “tres casos” del cuaderno)

Defino $(T_{\min}=\min(T_1,T_2)$) y $(T_{\max}=\max(T_1,T_2)$).

**Caso 1 — Sin traslape inicial (no superposición):**  
Para $(t<0$) todavía no se “tocan” los intervalos, así que:
$[
y(t)=0 \quad \text{(este es el primer caso que menciona el cuaderno).}
$]

**Caso 2 — Con traslape (intersección variable):**  
A medida que $(t$) avanza, la intersección crece, puede mantenerse constante (si $(T_1\neq T_2$)) y luego decrece:

- **Creciente:** $(0\le t\le T_{\min}$)  
  $[
  y(t)=t.
  $]
  (El solapamiento aumenta linealmente desde 0 hasta \(T_{\min}\).)

- **Meseta (solo si $(T_1\neq T_2$)):** $(T_{\min}\le t\le T_{\max}$)  
  $[
  y(t)=T_{\min}.
  $]
  (El solapamiento queda limitado por el rectángulo más corto.)

- **Decreciente:** $(T_{\max}\le t\le T_1+T_2$)  
  $[
  y(t)=T_1+T_2 - t.
  $]
  (El solapamiento disminuye linealmente hasta desaparecer.)

**Caso 3 — Sin traslape final:**  
Para $(t\ge T_1+T_2$) los intervalos ya se separaron por completo:
$[
y(t)=0.
$]

---

### 4) Resultado pieza a pieza (forma cerrada)
$[
y(t)=
\begin{cases}
0, & t<0,\\[4pt]
t, & 0\le t\le T_{\min},\\[4pt]
T_{\min}, & T_{\min}\le t\le T_{\max},\\[4pt]
T_1+T_2 - t, & T_{\max}\le t\le T_1+T_2,\\[4pt]
0, & t\ge T_1+T_2.
\end{cases}
$]

**Observación:** si $(T_1=T_2=T$), la “meseta” desaparece y queda un **triángulo isósceles**:
$[
y(t)=
\begin{cases}
0, & t<0,\\
t, & 0\le t\le T,\\
2T-t, & T\le t\le 2T,\\
0, & t\ge 2T.
\end{cases}
$]

---

### 5) Escalado por alturas (caso general)
Si las alturas no son 1 (por ejemplo $(x$) con altura $(A_1$) y $(h$) con altura $(A_2$)), el resultado se **multiplica por $(A_1A_2$)**:
$[
y_{\text{general}}(t) = A_1 A_2 \cdot y_{\text{unitario}}(t).
$]

---

### 6) Verificación numérica (opcional)
Con muestreo fino $(\Delta t$), la convolución discreta cumple:
$[
y_{\text{num}}[n]\ \approx\ \sum_k x[k]\;h[n-k]\;\Delta t,
$]
y debe **coincidir** con la curva analítica por tramos.  
(En el código de abajo hago esta verificación y grafico ambas curvas).


```python
import numpy as np
import matplotlib.pyplot as plt

# ----- parámetros (puedes cambiarlos) -----
T1 = 2.0   # ancho del primer rectángulo
T2 = 3.0   # ancho del segundo rectángulo
A1 = 1.0   # altura x(t)
A2 = 1.0   # altura h(t)  (si A1,A2 != 1, y(t) se escala por A1*A2)

# ----- eje temporal continuo (muestreo fino para visualizar) -----
dt = 1e-3
t = np.arange(-1.0, T1+T2+1.0, dt)  # tiempo para graficar

# ----- definiciones rectangulares causales -----
def rect_causal(t, T, A=1.0):
    """A*(u(t)-u(t-T)) -> 1 entre [0, T], 0 fuera."""
    return A * ((t >= 0) & (t <= T)).astype(float)

x = rect_causal(t, T1, A1)
h = rect_causal(t, T2, A2)

# ----- y(t) por fórmula de traslape (analítica pieza a pieza) -----
Tmin, Tmax = min(T1, T2), max(T1, T2)

y = np.zeros_like(t)
# Caso crec: 0 <= t <= Tmin -> y = t
mask1 = (t >= 0) & (t <= Tmin)
y[mask1] = t[mask1]

# Caso meseta (si T1 != T2): Tmin <= t <= Tmax -> y = Tmin
mask2 = (t >= Tmin) & (t <= Tmax)
y[mask2] = Tmin

# Caso decreciente: Tmax <= t <= T1+T2 -> y = T1 + T2 - t
mask3 = (t >= Tmax) & (t <= T1+T2)
y[mask3] = (T1 + T2) - t[mask3]

# Escala por alturas si A1,A2 != 1
y *= (A1 * A2)

# ----- verificación numérica por “convolución” discreta (CORREGIDA) -----
x_d = rect_causal(t, T1, A1)
h_d = rect_causal(t, T2, A2)

y_num = np.convolve(x_d, h_d) * dt           # *dt aproxima la integral
t_start = t[0] + t[0]                         # inicio del soporte (suma de los inicios)
t_num = t_start + np.arange(len(y_num)) * dt  # eje que coincide con y_num

# ----- gráficas -----
plt.figure(figsize=(11,7))
plt.subplot(3,1,1)
plt.plot(t, x, label='x(t) = rect(0..T1)')
plt.grid(True); plt.legend(); plt.ylabel('amplitud')

plt.subplot(3,1,2)
plt.plot(t, h, 'C1', label='h(t) = rect(0..T2)')
plt.grid(True); plt.legend(); plt.ylabel('amplitud')

plt.subplot(3,1,3)
plt.plot(t, y, 'C2', lw=2, label='y(t) (analítica por traslape)')
plt.plot(t_num, y_num, 'k--', alpha=0.6, label='y_num ≈ x*h (discreta)')
plt.grid(True); plt.legend(); plt.xlabel('t [s]'); plt.ylabel('y(t)')
plt.suptitle('Convolución de dos rectángulos causales — tres casos')
plt.tight_layout()
plt.show()
```

1: Comparación entre la salida de la EDO y la obtenida por convolución**
  
- **Compare la señal de salida obtenida al resolver la EDO con la obtenida mediante la convolución. ¿Son iguales?**

---


El sistema analizado en el cuaderno es un sistema **lineal e invariante en el tiempo (LTI)** descrito por la ecuación diferencial:

$[
\frac{dy(t)}{dt} + a\,y(t) = x(t),
$]

con condición inicial nula \(y(0^-)=0\).  
La función de transferencia en el dominio de Laplace es:

$[
H(s) = \frac{1}{s + a}.
$]

La **respuesta al impulso** del sistema es entonces:

$[
h(t) = e^{-a t}\,u(t),
$]

y la **salida por convolución** se obtiene como:

$[
y(t) = (h * x)(t) = \int_0^t e^{-a(t - \tau)}\,x(\tau)\,d\tau.
$]

---

### **2. Procedimiento**
1. Se simula el sistema usando la ecuación diferencial $(y'(t) + a y(t) = x(t)$) con una entrada $(x(t)$) (por ejemplo, un escalón o un pulso rectangular).  
2. Se calcula la salida equivalente aplicando la **convolución** de $(x(t)$) con $(h(t) = e^{-a t}u(t)$).  
3. Finalmente, se comparan ambas señales $(y(t)$) obtenidas por los dos métodos.

---

### **3. Observaciones**
- Para condiciones iniciales nulas ($(y(0^-)=0$)), las dos respuestas son **idénticas**, ya que la convolución reproduce exactamente la solución particular de la EDO.  
- La pequeña diferencia numérica entre ambas curvas solo se debe al paso de muestreo ($(\Delta t$)) usado en la integración.  
- La forma de $(y(t)$) depende de la entrada:
  - Si $(x(t) = u(t)$):  
    $[
    y(t) = \frac{1 - e^{-a t}}{a}.
    $]
  - Si $(x(t)$) es un pulso de duración $(T_p$): la salida crece exponencialmente mientras el pulso está activo y decae cuando termina.

---
La salida obtenida por **resolución directa de la EDO** y la obtenida mediante la **convolución con la respuesta al impulso** son **iguales** siempre que:
- El sistema sea **LTI** (lineal e invariante en el tiempo), y  
- Las **condiciones iniciales sean nulas**.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import lsim, TransferFunction

# -------- Parámetros del sistema (EDO: y' + a y = x) --------
a = 2.0                 # puedes cambiar a
dt = 1e-3
T  = 3.0
t  = np.arange(0, T+dt, dt)

# -------- Entradas de prueba --------
# 1) Escalón u(t)
u_step = np.ones_like(t)

# 2) Pulso rectangular de duración Tp (desde t=0 hasta t=Tp)
Tp = 0.6
u_pulse = (t >= 0) & (t <= Tp)
u_pulse = u_pulse.astype(float)

# -------- 1) Salida resolviendo la EDO (simulación de H(s)=1/(s+a)) --------
H = TransferFunction([1.0], [1.0, a])   # numerador, denominador
tout1, y_step_edo, _  = lsim(H, U=u_step,  T=t)     # a escalón
tout2, y_pulse_edo, _ = lsim(H, U=u_pulse, T=t)     # a pulso

# -------- 2) Salida por convolución con h(t)=e^{-a t} u(t) --------
h = np.exp(-a*t)                         # respuesta al impulso (u(t) implícito porque t>=0)
# Convolución continua aproximada: y ≈ (x * h) * dt
y_step_conv  = np.convolve(u_step,  h)[:len(t)]  * dt
y_pulse_conv = np.convolve(u_pulse, h)[:len(t)] * dt

# -------- Comparación numérica --------
err_step  = np.max(np.abs(y_step_edo  - y_step_conv))
err_pulse = np.max(np.abs(y_pulse_edo - y_pulse_conv))
print(f"Error máx (escalón): {err_step:.3e}")
print(f"Error máx (pulso)  : {err_pulse:.3e}")

# -------- Gráficas --------
plt.figure(figsize=(11,6))
plt.subplot(2,2,1)
plt.plot(t, u_step); plt.grid(True); plt.title("Entrada 1: u(t)")
plt.subplot(2,2,2)
plt.plot(t, u_pulse); plt.grid(True); plt.title(f"Entrada 2: pulso (Tp={Tp}s)")

plt.subplot(2,2,3)
plt.plot(t, y_step_edo, label="EDO");
plt.plot(t, y_step_conv, '--', label="Convolución")
plt.grid(True); plt.title("Salida a escalón"); plt.xlabel("t [s]"); plt.legend()

plt.subplot(2,2,4)
plt.plot(t, y_pulse_edo, label="EDO");
plt.plot(t, y_pulse_conv, '--', label="Convolución")
plt.grid(True); plt.title("Salida a pulso"); plt.xlabel("t [s]"); plt.legend()

plt.tight_layout(); plt.show()
```

2: Cálculo manual de la respuesta al impulso h(t)**


 -  **Comprobar manualmente la solución $( h(t) $) de la EDO cuando la entrada es $( x(t)=\delta(t) $). **
Recordar que $( \frac{d}{dt}\,\varepsilon(t) = \delta(t) $), donde $( \varepsilon(t) $) es la función escalón (Heaviside).

---

### **1. Ecuación diferencial base**
El sistema se describe por:
$[
\frac{dy(t)}{dt} + a\,y(t) = x(t),
$]
y para obtener la **respuesta al impulso**, hacemos:
$[
x(t) = \delta(t), \qquad y(t) = h(t).
$]

Sustituyendo:
$[
\frac{dh(t)}{dt} + a\,h(t) = \delta(t),
$]
con condición inicial nula \( h(0^-) = 0 \).

---

### **2. Resolución paso a paso**

#### a) Solución para \( t > 0 \)
Cuando $( t > 0 $), la delta desaparece ($(\delta(t)=0$)), así que la EDO homogénea es:
$[
\frac{dh(t)}{dt} + a\,h(t) = 0.
$]
Su solución general es:
$[
h(t) = C\,e^{-a t}, \quad t>0.
$]

#### b) Determinar la constante $(C$)
Integramos la EDO original alrededor de $( t=0 $):
$[
\int_{0^-}^{0^+} \left(\frac{dh(t)}{dt} + a\,h(t)\right) dt = \int_{0^-}^{0^+} \delta(t)\,dt.
$]

Evaluando los términos:
$[
[h(0^+) - h(0^-)] + a\int_{0^-}^{0^+}h(t)dt = 1.
$]

Como $(h(0^-)=0$) y el segundo término es despreciable (integral infinitesimal), queda:
$[
h(0^+) = 1.
$]

Entonces:
$[
C = 1.
$]

---

### **3. Resultado final**
$[
\boxed{h(t) = e^{-a t}\,u(t)}
$]

Es decir, la **respuesta al impulso** es un exponencial decreciente que comienza en 1 en $(t=0^+$).

---

### **4. Interpretación**
- $(h(t)$) describe cómo responde el sistema a un impulso unitario.  
- El parámetro $(a$) controla la **velocidad de decaimiento**:  
  - Si $(a>0$), el sistema es **estable** (la respuesta se atenúa).  
  - Si $(a=0$), el sistema es **integrador** (salida constante).  
  - Si $(a<0$), el sistema es **inestable** (crece exponencialmente).  
- Además, como $( \frac{d}{dt}\varepsilon(t) = $delta(t) $), también se puede comprobar que:
  $[
  h(t) = $(\frac{d}{dt}\big(e^{-a t}\varepsilon(t)\big )$,
  $]
  lo que confirma que la derivada del escalón exponencial produce la misma $(h(t)$).

---

La solución manual demuestra que la **respuesta al impulso del sistema** definido por  
$(\frac{dy(t)}{dt} + a\,y(t) = x(t)$)  
es $(h(t)=e^{-a t}u(t)$), coincidiendo con la obtenida mediante la transformada de Laplace y con la que usamos para la convolución en el subpunto anterior.

2.3 Demuestre si los siguientes sistemas de la forma \( y = H\{x\} \) son Sistemas Lineales e Invariantes en el Tiempo (SLIT)

---


Un sistema es **lineal** si cumple las dos propiedades:

1. **Homogeneidad:**  
   Si la entrada se multiplica por una constante, la salida también se multiplica por la misma constante.  
   $[
   \mathcal{H}\{\alpha x[n]\} = \alpha\,\mathcal{H}\{x[n]\}
   $]

2. **Aditividad:**  
   La respuesta a la suma de entradas es la suma de las respuestas individuales.  
   $[
   \mathcal{H}\{x_1[n] + x_2[n]\} = \mathcal{H}\{x_1[n]\} + \mathcal{H}\{x_2[n]\}
   $]

Y es **invariante en el tiempo (TI)** si al desplazar la entrada en el tiempo, la salida también se desplaza igual:
$[
\mathcal{H}\{x[n-n_0]\} = y[n-n_0]
$]

---

## **Sistema 1:**
$[
y[n] = \frac{1}{3}x[n] + 2x[n-1] - y[n-1]
$]

### **Análisis paso a paso**

**a) Linealidad:**

- No hay potencias, productos ni términos no lineales.  
- Si $(x_1[n] \to y_1[n]$) y $(x_2[n] \to y_2[n]$):  
  $[
  \mathcal{H}\{\alpha x_1 + \beta x_2\} = \frac{1}{3}(\alpha x_1+\beta x_2) + 2(\alpha x_1[n-1]+\beta x_2[n-1]) - (\alpha y_1[n-1]+\beta y_2[n-1])
  $]
  que es igual a $(\alpha y_1[n] + \beta y_2[n]$).  
  Por tanto, **es lineal**.

**b) Invariancia en el tiempo:**

Si desplazo la entrada \(x[n-n_0]\), la ecuación se convierte en:
\[
y[n-n_0] = \frac{1}{3}x[n-n_0] + 2x[n-1-n_0] - y[n-1-n_0],
\]
que tiene la misma forma.  
Por tanto, **sí es invariante en el tiempo.**


El sistema es **lineal e invariante en el tiempo (SLIT)**.  
Corresponde a un **sistema IIR** lineal con coeficientes constantes.

---

##  **Sistema 2:**
$[
y[n] = \sum_{k=-\infty}^{n} x^2[k]
$]

### **a) Linealidad**

- Contiene un término $(x^2[k]$), que rompe la propiedad de homogeneidad.  
- Si \(x[n]\) se multiplica por $(\alpha$):
  $[
  y_{\text{nuevo}}[n] = \sum (\alpha x[k])^2 = \alpha^2 \sum x^2[k] \neq \alpha y[n].
  $]
  ⇒ No cumple linealidad.

### **b) Invariancia en el tiempo**

Desplazo la entrada \(x[n-n_0]\):
$[
\mathcal{H}\{x[n-n_0]\} = \sum_{k=-\infty}^{n} x^2[k-n_0].
$]
Cambiando variable \(m = k - n_0\):
$[
\sum_{m=-\infty}^{n-n_0} x^2[m] = y[n-n_0].
$]
⇒ Sí cumple **invariancia en el tiempo**.


El sistema **no es lineal**, pero **sí es invariante en el tiempo**.  
Por tanto, **no es SLIT**.

---

## **Sistema 3:**
$[
y[n] = \operatorname{median}\{x[n-1], x[n], x[n+1]\}
$]

### **a) Linealidad**

- La función **mediana** no cumple ni homogeneidad ni aditividad.

**Ejemplo rápido:**
- $(x_1[n] = [0, 100, 0]$) → mediana = 0  
- $(x_2[n] = [0, 0, 100]$) → mediana = 0  
- $(x_1[n] + x_2[n] = [0, 100, 100]$) → mediana = 100  
Entonces:
$[
\mathcal{H}\{x_1+x_2\} \neq \mathcal{H}\{x_1\} + \mathcal{H}\{x_2\}.
$]
⇒ No es lineal.

### **b) Invariancia en el tiempo**

El operador mediana se aplica siempre sobre la misma ventana relativa a $(n$):  
si la entrada se desplaza, la ventana también se desplaza igual.  
Por tanto, **sí es invariante en el tiempo.**


**No lineal, pero sí TI.**  
**No es SLIT.**

---

## **4️Sistema 4:**
$[
y(t) = A\,x(t) + B,\quad A,B\in\mathbb{R}
$]

### **a) Linealidad**

- Si $(B=0$):  
  $[
  y(t) = A x(t)
  $]
  Cumple homogeneidad y aditividad ⇒ **lineal**.

- Si $(B\neq0$):  
  $(\mathcal{H}\{0\}=B\neq0$) ⇒ no cumple homogeneidad.  
  Por tanto, **no lineal**.

### **b) Invariancia en el tiempo**

Desplazo la entrada \(x(t-t_0)\):
$[
\mathcal{H}\{x(t-t_0)\} = A\,x(t-t_0) + B = y(t-t_0).
$]
⇒ Siempre **invariante en el tiempo**.


- Si \(B=0\) → **Sí es SLIT**.  
- Si \(B\neq0\) → **No es lineal**, aunque sigue siendo **TI**.

2.4 Hallar la salida $( y[n] $) de un sistema lineal e invariante en el tiempo (SLIT) ante la entrada:

$[
x[n] = \{-15,\, 5,\, -3^{\dagger},\, 0,\, 5,\, 7,\, -1\}
$]

con respuesta al impulso:

$[
h[n] = \{1,\, -2,\, 0^{\dagger},\, 1,\, -2\},
$]

donde el símbolo $( \dagger $) indica el valor correspondiente a $( n = 0 $).  
Luego, repetir el proceso para el sistema con respuesta al **escalón**:

$[
s[n] = \{-1,\, 6,\, -10,\, 3^{\dagger},\, 1,\, -10,\, 2,\, 5\}.
$]

---

### **1Parte 1 — Convolución con la respuesta al impulso ($h[n]$)**

El sistema es **lineal e invariante en el tiempo**, por lo tanto la salida se obtiene aplicando la **convolución discreta**:

$[
y[n] = (x * h)[n] = \sum_{k=-\infty}^{\infty} x[k]\,h[n-k].
$]

Cada valor de $(y[n]$) se calcula multiplicando punto a punto $(x[k]$) con la versión invertida y desplazada de $(h[k]$).

El origen de $(x[n]$) está en la posición 2 (el valor $(-3$))  
y el de $(h[n]$) en la posición 2 (el valor $(0$)).  
Por tanto, el origen de $(y[n]$) está en:

$[
n_y = n_x + n_h = 4.
$]

*2.5 — Sistemas en serie con señal Gaussiana*

Sea la señal:
$[
x(t)=e^{-a t^2}, \quad a>0
$]
El sistema **A** tiene la relación:
$[
y_A(t)=x^2(t)=e^{-2a t^2}.
$]
El sistema **B** es LTI con respuesta al impulso:
$[
h_B(t)=B\,e^{-b t^2}.
$]

Se pide:
1. **a)** Salida del sistema en serie $(x(t)\rightarrow h_B(t)\rightarrow y_A(t)\rightarrow y(t)$).  
2. **b)** Salida del sistema en serie $(x(t)\rightarrow y_A(t)\rightarrow h_B(t)\rightarrow y(t)$).

---



#### **Caso a)**
Primero se aplica el sistema $(B$) y luego el sistema $(A$).

$[
y(t)=y_A(t)=\big[(x*h_B)(t)\big]^2
$]
donde la convolución entre dos Gaussianas $(e^{-a t^2}$) y $(e^{-b t^2}$) produce otra Gaussiana:
$[
(x*h_B)(t)=C\,e^{-\frac{a b}{a+b}t^2},\quad C=\sqrt{\frac{\pi}{a+b}}\,B.
$]
Por tanto:
$[
y(t)=C^2\,e^{-\frac{2 a b}{a+b}t^2}.
$]

#### **Caso b)**
Primero se aplica el sistema \(A\) y luego \(B\):
$[
y(t)= (y_A*h_B)(t) = \int e^{-2a\tau^2}\,B e^{-b(t-\tau)^2}\,d\tau.
$]
El resultado es nuevamente Gaussiano:
$[
y(t)=B\sqrt{\frac{\pi}{2a+b}}\,e^{-\frac{2ab}{2a+b}t^2}.
$]

---

En ambos casos la salida es una función Gaussiana $(e^{-k t^2}$),  
pero el **parámetro de dispersión** cambia según el orden de los sistemas:
- En el **caso a)** la atenuación depende de $(\frac{2ab}{a+b}$).
- En el **caso b)** depende de $(\frac{2ab}{2a+b}$).

Para comprobarlo numéricamente se simulan las convoluciones en Python.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve

# Parámetros
a = 1.0
b = 0.8
B = 1.0
t = np.linspace(-5, 5, 1000)

# Señal de entrada y sistemas
x = np.exp(-a * t**2)
hB = B * np.exp(-b * t**2)

# --- Caso (a): x → hB → yA
conv_a = convolve(x, hB, mode='same') * (t[1]-t[0])
y_a = conv_a**2

# --- Caso (b): x → yA → hB
yA = x**2
conv_b = convolve(yA, hB, mode='same') * (t[1]-t[0])
y_b = conv_b

# --- Gráficas
plt.figure(figsize=(10,7))
plt.subplot(3,1,1)
plt.plot(t, x, label='x(t)=e^{-a t²}')
plt.title('Entrada Gaussiana y sistemas A y B')
plt.grid(True); plt.legend()

plt.subplot(3,1,2)
plt.plot(t, hB, label='h_B(t)=B e^{-b t²}', color='orange')
plt.grid(True); plt.legend()

plt.subplot(3,1,3)
plt.plot(t, y_a, label='Salida caso (a)', color='C2')
plt.plot(t, y_b, '--', label='Salida caso (b)', color='C1')
plt.grid(True); plt.legend()
plt.xlabel('t'); plt.tight_layout()
plt.show()
```
